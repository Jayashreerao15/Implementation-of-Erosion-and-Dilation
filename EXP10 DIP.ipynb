```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image1=np.zeros((100,500),dtype='uint8')
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(image1,'Jayashree Rao',(5,70),font,2,(255),5,cv2.LINE_AA)
plt.title("Original Image")
plt.imshow(image1,cmap='gray')
```

```python
kernel = np.ones((5,5),np.uint8)
kernel1=cv2.getStructuringElement(cv2.MORPH_CROSS,(7,7))
image_erode1=cv2.erode(image1,kernel1)
plt.title("ERODED IMAGE")
plt.imshow(image_erode1,cmap='gray')
```

```python
image_dilatel=cv2.dilate(image1,kernel1)
plt.title("DILATED IMAGE")
plt.imshow(image_erode1,cmap='gray')
```
